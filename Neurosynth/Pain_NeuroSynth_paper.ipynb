```python
import neurosynth as ns
ns.dataset.download(path='.', unpack=True)
from neurosynth.base.dataset import Dataset
dataset = Dataset('../current_data/database.txt')
dataset.add_features('../current_data/features.txt')
dataset.get_feature_names()
ids = dataset.get_studies(features='pain*', frequency_threshold=0.001)
len(ids)
from neurosynth.analysis import meta
ma = meta.MetaAnalysis(dataset, ids)
ma.save_results('../Neurosynth/pain')
print "Found %d studies." % len(ids)
```
